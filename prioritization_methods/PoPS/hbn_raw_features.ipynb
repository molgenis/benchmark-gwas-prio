## rename the columns of gene expression file

```python
import pandas as pd
import os
from pathlib import Path

file = "C:\hbn_average_expression.txt"

data = pd.read_csv(file, sep='\t', header=0)
```

```python
os.path.basename(file)
```

```python
file_name = Path(file).stem
file_name
```

```python
data
```

```python
#hbn_average_expression

new_cols = [file_name + "." + column for column in data.columns[1:-1].values]
new_cols.insert(0, data.columns[0])
new_cols.insert(len(data.columns), data.columns[-1])
```

```python
# data.columns[1:-1].values = new_cols
new_cols
```

```python
data.columns = new_cols
data
```

```python
out_file = Path("C:\\Users\\Gebruiker\\OneDrive\\Documents\\DSLS\\integrated_omics\\project\\POPS") / "hbn_average_expression.txt"

data.to_csv(out_file, sep="\t", index = False)
```
